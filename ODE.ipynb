```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
# function that returns dz/dt
def dydx(y,x,c0):
    v0=0.5#0.707
    k=0.1 #determines speed of fluid
    dydx = c0*v0/math.sqrt(1-pow(c0,2)*(pow(v0,2)-k*pow(x,2))) # linear flow 
    return dydx

# initial condition
y0=0

# time points
x = np.linspace(0.,1.75,100)
c0=1.5453562
# solve ODE
y = odeint(dydx,y0,x,args=(c0,))

# plot results
plt.plot(x,y,'b-',label='y(x)')
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='best')
plt.show()
```
